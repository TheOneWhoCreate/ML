```python
import numpy as np
import pandas as pd
import math
from sklearn import linear_model
```

```python
x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])
```

```python
m_curr = b_curr = 0
iteration = 1000
n = len(x)
learning_rate = 0.08
for i in range(iteration):
  y_pred = m_curr * x + b_curr
  md = -(2/n)*sum(x*(y-y_pred))
  bd = -(2/n)*sum(y-y_pred)
  m_curr = m_curr - learning_rate * md
  b_curr = b_curr - learning_rate * bd
  cost = (1/n)*sum([val**2 for val in (y-y_pred)])
```

```python
testScore = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Csv_file/test_scores.csv')
testScore.keys()
```

```python
testScoreDf = pd.DataFrame(testScore)
x1 = np.array(testScoreDf['math'])
y1 = np.array(testScoreDf['cs'])
```

```python
m_curr1 = 0
b_curr1 = 0
ita = 1000000
learningRate = 0.0002
n1 = len(x1)
cost_previous = 0

for i in range(ita):
  y_pred1 = m_curr1 * x1 + b_curr1

  md1 = -(2/n1) * sum(x1 * (y1 - y_pred1))
  bd1 = -(2/n1) * sum(y1 - y_pred1)

  m_curr1 = m_curr1 - learningRate * md1
  b_curr1 = b_curr1 - learningRate * bd1

  cost1 = (1/n1) * sum([val**2 for val in (y1 - y_pred1)])
  if math.isclose(cost1, cost_previous, rel_tol=1e-15):
    break
  cost_previous = cost1
```

```python
reg = linear_model.LinearRegression()
reg.fit(testScoreDf.drop(columns= ['name', 'cs']), testScoreDf['cs'])
m = reg.coef_
b = reg.intercept_
```
