```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
homeprices = {
    "area": [2600, 3000, 3200, 3600, 4000],
    "prices": [550000, 565000, 610000, 680000, 725000]
}

homePricesDf = pd.DataFrame(homeprices)
print(homePricesDf.head())
```

```python
%matplotlib inline
plt.xlabel('Area')
plt.ylabel('Price')
plt.scatter(homePricesDf.area, homePricesDf.prices, color='red')
```

```python
reg = linear_model.LinearRegression()
reg.fit(homePricesDf[['area']], homePricesDf[['prices']])
```

```python
%matplotlib inline
plt.xlabel('Area', fontsize= 20)
plt.ylabel('Price', fontsize= 20)
plt.scatter(homePricesDf.area, homePricesDf.prices, color='red')
plt.plot(homePricesDf.area, reg.predict(homePricesDf[['area']]), color= "blue")
```

```python
reg.predict(pd.DataFrame({'area': [5000]}))
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
areas = [1000, 1500, 2300, 3540, 4120, 4560, 5490, 3460, 4750, 2300, 900, 8600, 7100]
areaDf = pd.DataFrame({'area': areas})
predictedPrice = reg.predict(areaDf)
areaDf["prices"] = predictedPrice
print(areaDf)
```
