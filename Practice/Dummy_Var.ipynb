```python
import pandas as pd
import numpy as np
from sklearn import linear_model
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
```

```python
homepricestown = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Csv_file/homepricestown.csv')
homepricestown.keys()
print(homepricestown)
```

```python
dummies = pd.get_dummies(homepricestown['town'])
mergerd = pd.concat([homepricestown, dummies], axis='columns')
```

```python
finalDf = mergerd.drop(columns= ['town', 'monroe township'])
townPriceDf = pd.DataFrame(finalDf)
print(townPriceDf)
```

```python
model = linear_model.LinearRegression()
```

```python
X = townPriceDf.drop(columns= ['price'])
Y = townPriceDf['price']

model.fit(X, Y)
```

```python
model.predict(pd.DataFrame({
    'area': [2800, 3400, 3800],
    'robinsville': [1, 0, 0],
    'west windsor': [0, 1, 0]
}))
```

```python
model.score(X, Y)
```

```python
X = homepricestown.drop(columns= ['price'])
Y = homepricestown['price']
```

```python
preprocessor = ColumnTransformer(
    transformers= [('encoded', OneHotEncoder(drop='first', sparse_output=False), ['town'])
    ],
    remainder='passthrough',
    force_int_remainder_cols=False
)

pipeline = Pipeline([
    ('preprocessing', preprocessor),
    ('model', linear_model.LinearRegression())
])

pipeline.fit(X,Y)
```

```python
pipeline.predict(pd.DataFrame({
    'town': ['west windsor', 'robinsville'],
    'area': [3400, 2800]
}))
```

```python
pipeline.score(X, Y)
```

```python
carPrices = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Csv_file/carprices.csv')
carPrices.keys()
print(carPrices)
```

```python
X = carPrices.drop(columns= ['Sell Price($)'])
Y = carPrices['Sell Price($)']
```

```python
process = ColumnTransformer(
    transformers=[
        ('car', OneHotEncoder(drop= 'first', sparse_output = False), ['Car Model'])
    ],
    remainder='passthrough'
)
pl = Pipeline([
    ('processing', process),
    ('model', linear_model.LinearRegression())
])

pl.fit(X, Y)
```

```python
pl.predict(pd.DataFrame({
    'Car Model': ['BMW X5', 'Mercedez Benz C class', 'BMW X5'],
    'Mileage': [10000,  45000, 86000],
    'Age(yrs)': [6, 4, 7]
}))
```

```python
pl.score(X, Y)
```
