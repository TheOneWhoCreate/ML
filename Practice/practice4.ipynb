```python
import pandas as pd
import numpy as np
from word2number import w2n
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
#Reading Data
homePrices = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Csv_file/homeprices.csv')
homePrices.keys()
```

```python
homePricesDf = pd.DataFrame(homePrices)
print(homePricesDf)
```

```python
homePricesDf.info()
```

```python
homePricesDf.describe()
```

```python
homePricesDf.isnull().sum()
```

```python
homePricesDf['bedrooms'] = np.floor(
    homePricesDf['bedrooms'].fillna(homePricesDf['bedrooms'].median())
)
print(homePricesDf)
```

```python
reg = linear_model.LinearRegression()
reg.fit(homePricesDf.drop(columns= ['price']), homePricesDf.price)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict(pd.DataFrame({
    'area': [3200, 2500],
    'bedrooms': [4, 4],
    'age': [18, 5]
}))
```

```python
hiring = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Csv_file/hiring.csv')
hiring.columns = ['experience', 'test_score', 'interview_score', 'salary']
hiring.keys()
```

```python
hiringDf = pd.DataFrame(hiring)
print(hiringDf)
```

```python
hiringDf.isnull().sum()
```

```python
hiringDf['experience'] = hiringDf['experience'].fillna("Zero")
hiringDf['experience'] = hiringDf['experience'].apply(
    w2n.word_to_num
)

hiringDf['test_score'] = np.floor(
    hiringDf['test_score'].fillna(hiringDf['test_score'].median())
)

print(hiringDf)
```

```python
reg = linear_model.LinearRegression()
reg.fit(hiringDf.drop(columns= ['salary']), hiringDf['salary'])
```

```python
salPredict= reg.predict(pd.DataFrame({
    'experience': [2, 12],
    'test_score': [9, 10],
    'interview_score': [6, 10]
}))

print(np.round(salPredict))
```
