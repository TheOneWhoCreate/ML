<font color="turquoise"><h3>Generating dataset</h3>

```python
#Importing modules
import numpy as np
import pandas as pd
```

```python
students = pd.DataFrame({
    "student_id": [101, 102, 103, 104, 105, 106, 107, 108],
    "name": ["Amit", "Riya", "John", "Sara", "Karan", "Meena", "Arjun", "Nisha"],
    "age": [20, 21, 19, np.nan, 22, 20, 23, 21],
    "math_score": [85, 90, np.nan, 88, 76, 95, 89, 92],
    "science_score": [78, np.nan, 80, 85, 70, 96, 88, np.nan],
    "city": ["Kolkata", "Delhi", "Mumbai", "Delhi", "Kolkata", "Mumbai", "Delhi", "Kolkata"]
})

students
```

```python
attendance = pd.DataFrame({
    "student_id": [101, 102, 103, 104, 105, 106, 107, 109],
    "attendance_percentage": [92, 85, 88, 90, np.nan, 95, 95, 75]
})

attendance
```

<font color= "skyblue"><h3>1. Reshaping the data</h3>

```python
#creating a array with toal score from the student table
score = np.array(students[["math_score", "science_score"]])

print("Original shape: ", score.shape)
#Reshaping the data
score_reshaped = score.reshape(16, 1)
print("Reshaped data: ", score_reshaped.shape)
```

<font color= 'skyblue'><h3>2. Filtering the data</h3>

```python
#Filtering on single condition
single_filtered_df = students[students['city'] == 'Delhi']
print("Students from delhi: \n")
print(single_filtered_df[['student_id', 'name', 'city']])
print("\n")

#filtering on multiple condition
multi_filtered_df = students[
    (students['city'] == 'Delhi') &
    (students['age'] > 20)]

print("Students from delhi and age avobe 20: \n")
print(multi_filtered_df[['student_id', 'name', 'city', 'age']])
```

<font color= 'skyblue'><h3>3. Merging the data</h3>

```python
#Merging student table and attendance table
mrg_df = pd.merge(students, attendance, on='student_id', how='inner')

print("Merged data: \n")
print(mrg_df[['student_id', 'name', 'attendance_percentage']])
```

<font color= 'skyblue'><h3>4. Handling missing values</h3>

```python
#Unhandled missing values
print("Raw data: \n")
print(students)
print("\n")

#Handling missing values
students["age"]= students["age"].fillna(
    students["age"].mean())
students["math_score"]= students["math_score"].fillna(
    students["math_score"].mean())
students["science_score"]= students["science_score"].fillna(
    students["science_score"].mean())

print("Processed data: \n")
print(students)
```

<font color= 'skyblue'><h3>5. Feature Normalization (Min-Max normalization)</h3>

```python
print("Feature Normalization: \n")
norm_df = students.copy()
col= norm_df["math_score"]

norm_df['math_score_norm']= (col - col.min()) / (col.max() - col.min())
print(norm_df)
```
