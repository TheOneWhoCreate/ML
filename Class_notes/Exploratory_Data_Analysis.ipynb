## <font color="Blue"> **2.3 Prepare a Dataframe taking Features and Target:** </font>

```python
# Let's convert the iris Bunch object to a dataframe
Iris_Df = pd.DataFrame(data= np.c_[iris['data'], iris['target']],
                       columns= iris['feature_names'] + ['species'])
##Note: The function of np.c is to concatenate multiple NumPy arrays
##      column-wise to form a single array suitable for DataFrame creation.
## Alternately we can use:
## iris_df = pd.DataFrame(np.concatenate((iris['data'], iris['target'].reshape(-1, 1)),axis=1))


#Replace the numerical target values with class label names under 'species' for better interpretation
Iris_Df.species = np.where(Iris_Df.species == 0.0, 'setosa',
                        np.where(Iris_Df.species == 1.0, 'versicolor', 'virginica'))

#Let's remove spaces from column name
Iris_Df.columns = Iris_Df.columns.str.replace(' ','')

#Check the prepared dataframe
Iris_Df.head(-5)
```

## <font color="blue"> **3.1.2 Histogram with KDE curve (Distribution Analysis):**</font>

Pandas does not provide a built-in method to draw histograms with KDE Curve (Better Insight) for all features together. Seaborn’s histplot() function is used inside a loop to visualize feature distributions with KDE efficiently.
Different colors can be assigned to multiple subplots using a color list or Seaborn color palette to enhance clarity and visual appeal.

```python
import seaborn as sns
import matplotlib.pyplot as plt

#Select data types of the features
features = Iris_Df.select_dtypes(include='number').columns

#Define four different colors
colors = ['blue', 'green', 'red', 'violet']

plt.figure(figsize=(8, 6))

##Loop to plot all the features's histogram with KDE
for i, (feature, color) in enumerate(zip(features, colors), 1):
    plt.subplot(2, 2, i)
    sns.histplot(Iris_Df[feature], kde=True, color=color)
    plt.title(f'Distribution of {feature}')

plt.tight_layout()
plt.show()
```

## <font color="magenta"> **Key Insight from Histogram with KDE:**</font>
* Sepal length varies smoothly across samples and shows limited class distinction.
* Sepal width distribution is fairly symmetric and narrow and hence it is less informative for class discrimination.
* Petal length has clear bimodal distribution and hence strongly reflects class structure and is highly discriminative.
* Petal width also shows bimodal or multimodal distribution. It' peaks correspond to different species groups and hence provides strong evidence of class separability.

## <font color="blue"> **3.1.3 Overlaid Histograms:**</font>
Overlaid histograms are used to compare the distribution of numerical features across different classes in a dataset.

* Each subplot shows distribution of one feature
* Different colors represent different species
* Overlapping bars help compare class-wise distributions
* This visualization clearly shows which features separate classes well

```python
##Overlaid Histogram visualization
import matplotlib.pyplot as plt
import seaborn as sns


#Select data types of the features
features = Iris_Df.select_dtypes(include='number').columns

##Set titles of the histograms:
titles = ['Histogram of Sepal Length',
          'Histogram of Sepal Width',
          'Histogram of Petal Length',
          'Histogram of Petal Width'
        ]

##Declare figure size
plt.figure(figsize=(9, 6))

##Plot overlaid histograms for each features one by one
for i, (feature, title) in enumerate(zip(features, titles), 1):
    plt.subplot(2, 2, i)
    sns.histplot(
        data=Iris_Df,
        x=feature,
        hue='species',
        multiple='layer',     # overlapping bars
        bins=20,
        edgecolor='black',
        alpha=0.7
    )
    plt.title(title, fontsize=12)
    plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

# <font color="magenta"> **Key Observations:**</font>
* By using petal length or petal width we can separate iris-setosa.
* By using sepal length, sepal width we can’t do anything because it’s all messed up and we can’t separate the flowers.
* In petal width iris setosa is not distributed properly.
* Petal length can be used as a differentiating factor in terms of the distribution of the 3 flower species.

## <font color="blue"> **3.2.2 Box Plot of Individual Features for the Species:**</font>
Box plots are effective tools for comparing the distribution of numerical features across multiple categories. In the following example, box plots are used to analyze the variation of each feature of the Iris dataset among the three species. This visualization helps in identifying differences in central tendency, spread, and the presence of outliers.

```python
##Box Plot of Individual Features for the Species
import matplotlib.pyplot as plt

##Take the feature names of the data set
features = ['sepallength(cm)', 'sepalwidth(cm)',
            'petallength(cm)', 'petalwidth(cm)']

#Set the titles of the individual plot
titles = ['Box Plot of Sepal Length',
          'Box Plot of Sepal Width',
          'Box Plot of Petal Length',
          'Box Plot of Petal Width'
        ]
#Declare the colors to use for box plot of individual species
colors = ['gold', 'green', 'red']

#Declare figure size
plt.figure(figsize=(8, 8))

##Loop to plot histogram for each individual feature
for i, (feature, title) in enumerate(zip(features, titles), 1):
    plt.subplot(2, 2, i)

    #Select feature values of all classes for a particular feature
    data = [Iris_Df[Iris_Df['species'] == 'setosa'][feature],
            Iris_Df[Iris_Df['species'] == 'versicolor'][feature],
            Iris_Df[Iris_Df['species'] == 'virginica'][feature]]
    #Perform box-plot
    bp = plt.boxplot( data,
                      notch=True,           #Creates the notched shape
                      patch_artist=True,    #To fill boxes with colors
                      medianprops=dict(color='black', linewidth=2), #Median line color and width
                      tick_labels=['setosa', 'versicolor', 'virginica'] #Labels of boxes along x-axis
                    )
    #Set patch colors to the boxes
    for patch, color in zip(bp['boxes'], colors):
        patch.set_facecolor(color)

    plt.title(title, fontsize=14)
    plt.xlabel('Species')
    plt.ylabel(feature)

plt.tight_layout()
plt.show()
```

# <font color="magenta"> **Key Observations:**</font>

From the above box plot it’s clear sighted that:
* The Sepal Length for virginica and Sepal Width of Setosa and virginica both have outliers (They are the dots that out run the whiskers).
* The petal Length for setosa and versicolor and petal width for Setosa have outliers.
* While all the other boxplots looked almost balanced, we can see that that petal width for both setosa and versicolor are positively skewed as the median lie at the lower end of the boxplot.

The violin plot visualizes the distribution of all features —across all species in a single diagram. Each violin combines a box plot with a kernel density estimate, enabling simultaneous analysis of data spread, central tendency, and distribution shape.

**Purpose:**

# <font color="green"> **4. Bivariate Analysis:**</font>
**Bivariate analysis** is a statistical method used to explore the relationship between two variables. The goal is to understand whether and how the two variables are related — and if they are, then describe the nature, strength, and direction of that relationship.

**Cases between two variables and Plot:**

* **Numerical and Numerical:**  In this type, both the variables of bivariate data, independent and dependent, are having numerical values. For this case **Scatter plot** is used.

* **Numerical and Categorical:** When one variable is numerical and one is categorical- Bivariate Box Plot is used.

* **Categorical and Categorical:** When both the variables are categorical- Mosaic Plot is used.

## <font color="blue"> **4.1 Bivariate Scatter plot: Numerical Vs. Numerical:**</font>

```python
##Bivariate Scatter plot: Numerical Vs. Numerical
import matplotlib.pyplot as plt
import seaborn as sns

# Create a figure with two sub-figures side-by-side:
plt.figure(figsize=(8, 4))

# ----- 1st subplot: Sepal Length vs. Sepal Width ---------
plt.subplot(1, 2, 1)
sns.scatterplot(
    x='sepallength(cm)',
    y='sepalwidth(cm)',
    hue='species',
    data=Iris_Df
      )
plt.title('Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend(title='Species')

# ----- 2nd subplot: Petal Length vs. Petal Width ---------
plt.subplot(1, 2, 2)
sns.scatterplot(
    x='petallength(cm)',
    y='petalwidth(cm)',
    hue='species',
    data=Iris_Df
)
plt.title('Petal Length vs Petal Width')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.legend(title='Species')

# Adjust layout
plt.tight_layout()
plt.show()
```

# <font color="magenta"> **Analysis:**</font>

From the plot of **Sepal Length vs Sepal Width**, we can see that:
* The three species overlap significantly
* No clear boundary between versicolor and virginica
* Weak to moderate relationship between sepal length and sepal width

**Conclusion:**
Sepal length and sepal width together do not provide strong class separability and are less effective features for classification.

From the plot of **Petal Length vs Petal Width**, we can observe that:

* Setosa forms a completely separate cluster
* Versicolor and Virginica show partial overlap, but are mostly distinguishable
* Strong positive relationship between petal length and petal width

**Conclusion:**
Petal features exhibit high discriminative power and are highly suitable for machine learning classification tasks.

# <font color="green"> **5. Multivariate Analysis:**</font>

**Multivariate Analysis** is an extension of bivariate analysis which means it involves multiple variables at the same time to find correlation between them. Multivariate Analysis is a set of statistical model that examine patterns in multidimensional data by considering at once, several data variable.

**Common techniques used in multivariate analysis include:**
* Grouped Bar Chart
* Pair Plots (for pairwise views)
* Heatmaps (for correlation matrices)
* Faceted Grids (for multi-dimensional insights)
* Principal Component Analysis (PCA) for dimensionality reduction

## <font color="blue"> **5.1 Grouped Bar Chart:**</font>

**Grouped Bar Chart** can be useful for multivariate descriptive analysis. It is performed to compare multiple features across different species. The grouped bar chart visualizes the mean measurements of each feature, enabling easy comparison of species-wise variations and identification of discriminative attributes.

```python
##Grouped Bar Chart
#Compute the mean for each column by species
Iris_Df.groupby(by = "species").mean()

#Plot for mean of each feature for each label class
Iris_Df.groupby(by = "species").mean().plot(kind="bar")
plt.title('Class vs Measurements')
plt.ylabel('Mean measurement(cm)')
plt.xticks(rotation=0) # manage the xticks rotation
plt.grid(True)

# Use bbox_to_anchor option to place the legend outside plot area to be tidy
plt.legend(loc="upper left", bbox_to_anchor=(1,1))
```

## <font color="magenta"> **Key Insight:**</font>
* Clear Species Separation Using Petal Features
* Setosa Is Well-Separated
* Virginica Has the Largest Measurements
* Sepal Width Is Less Discriminative

## <font color="blue"> **5.2 Pair Plot Analysis:**</font>
**Pair plots** are effective tools for multivariate exploratory data analysis. They simultaneously display bivariate relationships and univariate distributions (on the diagonal), enabling comprehensive understanding of feature interactions and class separability, respectively.

```python
##Perform pair plot using sns.pairplot()
sns.pairplot(Iris_Df,  hue="species")
```

* Sepal width for setosa is high and length is low.
* Versicolor have average values for for sepal dimensions.
* Virginica has small width but large sepal length

The diagonal plots show the probability density function(PDF) with each feature as a variable and corresponding kernel density plot on Y-axis. These inddicate:

## <font color="blue"> **5.3 Heatmap for Correlation analysis:**</font>
**Correlation analysis** is an important step in exploratory data analysis to measure the degree of association between features. A correlation heat map provides a compact and intuitive visualization of correlation coefficients, enabling quick identification of strong positive and negative relationships among variables.

**Purpose:**
* Identify strongly correlated features
* Useful for feature selection

```python
##Correlation analysis by heatmap
plt.figure(figsize=(6, 5))

corr = Iris_Df.corr(numeric_only=True)

sns.heatmap(
    corr,
    annot=True,
    cmap='coolwarm',
    fmt=".2f"
)

plt.title("Correlation Heatmap of Iris Features")
plt.show()
```

## <font color="magenta"> **Observations:**</font>

The correlations among the features are visualised more easily on the above heatmap. The large amount of red and orange squares show that the majority of measurements are highly correlated with one another, except for sepal width.

**Key facts are:**

* Petal length and width are very highly positively correlated (r = 0.96), which tells us that as one gets larger so does the other, indicating that petal length and width have a close relationship.
* On the other hand, sepal measurements have a very weak relationship with one another (r = -0.11).
* Both petal length and width have very strong positive correlations with sepal length (r = 0.87 and 0.82 respectively) indicating that as both get larger, so does sepal length.
* Both petal length and width have fairly weak negative correlations with sepal width (r = -0.42 and -0.36 respectively) indicating there is not much of a relationship between these features.
* These correlations tell us that sepal width is not moderately or highly correlated with any other measurement.
