# <font color="green">**1. Introduction: <font color="green">**

**Decision Tree** and **Random Forest** are supervised machine learning algorithms used for **classification** and **regression** tasks.

* Decision Tree is a **non-parametric model** that splits data based on feature values.

* Random Forest is an **ensemble method** that combines multiple decision trees to improve generalization.

These algorithms are widely used due to:

* Interpretability (Decision Tree)

* High accuracy and robustness (Random Forest)

# <font color="green">**2. Decision Tree Classifier <font color="green">**

## <font color="blue">**2.1 Basic Idea <font color="blue">**

A Decision Tree recursively partitions the feature space into regions by selecting optimal splitting features based on impurity measures.

At each node:

* Select the best feature

* Determine the optimal threshold

* Split data into subsets

* Repeat recursively

# <font color="green">**3. Impurity Measures <font color="green">**

To decide the best split, we measure impurity (uncertainty) of a node.

A pure node contains samples of only one class.

## <font color="blue">**3.1 Entropy <font color="blue">**
Entropy measures randomness or disorder in the data.

For a dataset with $c$ classes:

$𝐻(𝑆)=\sum_{i=1}^{c} p_{i}log_2 p_{i}$

Where: $p_i$= probability of class $i$.

## Derivation of Entropy:
Entropy comes from Information Theory.

If an event has probability $𝑝$, its information content is:

$ I(p)= -log_2 p $

Expected information (entropy) is:

$𝐻(𝑆)=\sum p_{i}I(p_{i})$

$𝐻(𝑆)=-\sum p_{i}log_2 p_{i}$

## Entropy Properties:

If all samples belong to one class: $H=0$

For binary classification (maximum impurity): $H=1$ when $p_1=p_2=0.5$.

# <font color="blue">**3.2 Gini Index <font color="blue">**

Gini Index measures probability of misclassification.

$Gini(S)= 1-\sum_{i=1}^{c} p_i^2$

## Interpretation:

If we randomly label a sample based on class distribution,
Gini represents probability of incorrect classification.

For binary case:

$Gini= 1-(p^2 - (1-p)^2) = 2p(1-p)$
Maximum impurity occurs at $p=0.5$.

# <font color="green">**4. Information Gain <font color="green">**

Information Gain measures reduction in entropy after splitting.

$IG(S,A)= H(S)- H(S|A)$

where $H(S \mid A) = \sum_{v \in \text{values}(A)} \frac{|S_v|}{|S|} \, H(S_v)$

Thus, $IG$=Parent Entropy - Weighted Child Entropy

The feature with highest Information Gain is selected.

# <font color="green">**5. Splitting Criterion in Decision Tree <font color="green">**

At each node:

1. Compute impurity (Entropy or Gini).

2. Try all possible splits.

3. Calculate impurity after split.

4. Choose split that minimizes impurity.

Tree grows until:

* Maximum depth reached

* Minimum samples per node reached

* Node becomes pure

# <font color="green">**6. Overfitting in Decision Tree <font color="green">**

A fully grown tree:

* Low bias

* High variance

* Prone to overfitting

## Control parameters:

* max_depth

* min_samples_split

* min_samples_leaf

* pruning

# <font color="green">**7. Random Forest Classifier <font color="green">**

Random Forest is an ensemble of multiple decision trees.

## <font color="blue">**7.1 Motivation <font color="blue">**

Single Decision Tree:

* High variance

* Sensitive to small data changes

Random Forest reduces variance using:

1. Bagging (Bootstrap Aggregation)

2. Random feature selection

## <font color="blue">**7.2 Algorithm of Random Forest <font color="blue">**

For $B$ trees:

For each tree:

1. Draw bootstrap sample from training data.

2. At each split:

* Select random subset of features

* Choose best split among those features

3. Grow full tree (without pruning)

## Final prediction:

* Classification → Majority Voting

* Regression → Averaging

# <font color="magenta"> **Step 1: Import Required Libraries**

```python
##Import Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import load_breast_cancer

from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (accuracy_score, precision_score, recall_score,
    f1_score, roc_auc_score, roc_curve, confusion_matrix)
```

# <font color="magenta"> **Step 2: Load Dataset and Split (train: test=70:30)**

```python
##Load dataset and check the available keys in the dataset
##Separate Features and Targets. Check number of samples, features, class labels, etc.
## Perform stratified train-test split (70% train, 30% test)
##~~~~~Write codes below~~~~~~~~~~
data = load_breast_cancer()

X = data.data
y = data.target

print("Number of samples:", X.shape[0])
print("Number of features:", X.shape[1])
print("Class labels:", np.unique(y))

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, stratify=y, random_state=42)
```

# <font color="magenta"> **Step 3: Grid Search for Decision Tree**

## Find optimal parameters for the decision tree by gridsearch method.
Search the parametrs within the range as specified below:

* max_depth: None, 3, 5, 7, 10

* min_samples_split: 2, 5, 8, 10

* min_samples_leaf: 1, 2, 4, 6

* criterion: 'gini' and 'entropy'

```python
##Find out optimal parameters for DT classifiers using GridSearchCV() function
##Print the best DT parameters using '.best_params_' of the learned models
##Print the best DT CV accuracy using '.best_score_' of the learned models
##~~~~~Write codes below~~~~~~~~~~
param_grid_dt = {
    'max_depth': [None, 3, 5, 7, 10],
    'min_samples_split': [2, 5, 8, 10],
    'min_samples_leaf': [1, 2, 4, 6],
    'criterion': ['gini', 'entropy']
}

dt = DecisionTreeClassifier(random_state=42)

grid_dt = GridSearchCV(dt, param_grid_dt, cv=5, scoring='accuracy')
grid_dt.fit(X_train, y_train)

print("Best DT Parameters:", grid_dt.best_params_)
print("Best DT CV Accuracy:", grid_dt.best_score_)

Optimal_DT_Params = grid_dt.best_params_
```

```python
##Train DT classifier with optimal parameters obtained by grid search CV (**Optimal_DT_Params) and
##Predict class labels and probability score of Test set by the trained DT classifier
##~~~~~Write codes below~~~~~~~~~~
dt_model = DecisionTreeClassifier(**Optimal_DT_Params)
dt_model.fit(X_train, y_train)

y_pred_dt = dt_model.predict(X_test)
y_prob_dt = dt_model.predict_proba(X_test)[:,1]
```

```python
##Plot the learned DT classifier, show and save in drive
## Plot the using tree.plot_tree() function
##~~~~~Write codes below~~~~~~~~~~
plt.figure(figsize=(20,10))
tree.plot_tree(dt_model,
               feature_names=data.feature_names,
               class_names=data.target_names,
               filled=True)

plt.title("Decision Tree Classifier")
plt.show()
```

# <font color="magenta"> **Step 4: Grid Search for Random Forest Classifier**

## Find optimal parameters for the random forest model by gridsearch method.
Search the parametrs within the range as specified below:
* n_estimators: 50, 100, 150, 200, 300
* max_depth: None: 5, 10, 15
* min_samples_split: 2, 5
* min_samples_leaf: 1, 2
* max_features: 'sqrt', 'log2'

```python
##Find out optimal parameters for RF classifier using GridSearchCV() function
##Print the best parameters using '.best_params_' of the learned models
##Print the best DT CV accuracy using '.best_score_' of the learned models
##~~~~~Write codes below~~~~~~~~~~
param_grid_rf = {
    'n_estimators': [50,100,150,200,300],
    'max_depth': [None,5,10,15],
    'min_samples_split': [2,5],
    'min_samples_leaf': [1,2],
    'max_features': ['sqrt','log2']
}

rf = RandomForestClassifier(random_state=42)

grid_rf = GridSearchCV(rf, param_grid_rf, cv=5, scoring='accuracy')
grid_rf.fit(X_train, y_train)

print("Best RF Parameters:", grid_rf.best_params_)
print("Best RF CV Accuracy:", grid_rf.best_score_)

Optimal_RF_Params = grid_rf.best_params_
```

# <font color="magenta"> **Step 5: Predictions using tained RF model**

```python
##Train RF classifier with optimal parameters obtained by grid search CV
##Predict class labels and probability score of Test set by the trained RF classifier
##~~~~~Write codes below~~~~~~~~~~
rf_model = RandomForestClassifier(**Optimal_RF_Params, random_state=42)
rf_model.fit(X_train, y_train)

y_pred_rf = rf_model.predict(X_test)
y_prob_rf = rf_model.predict_proba(X_test)[:,1]
```

# <font color="magenta"> **Step 6: Compare performances of the two models:**

```python
##Write a function to evalute model performance metrics for accuracy, precision, recall, F1-score and roc_auc
##The function should have input : y_test; y_pred and y_prob of the respective models
##Calculate performaces of the two models
##Form a dataframe to compare the model performance metrics by DT and RF
## Print the dataframe to display the result
##~~~~~Write codes below~~~~~~~~~~
def evaluate_model(y_test, y_pred, y_prob):
  acc = accuracy_score(y_test, y_pred)
  prec = precision_score(y_test, y_pred)
  rec = recall_score(y_test, y_pred)
  f1 = f1_score(y_test, y_pred)
  roc = roc_auc_score(y_test, y_prob)

  return acc, prec, rec, f1, roc


dt_metrics = evaluate_model(y_test, y_pred_dt, y_prob_dt)
rf_metrics = evaluate_model(y_test, y_pred_rf, y_prob_rf)

metrics_df = pd.DataFrame({
  "Metric":["Accuracy","Precision","Recall","F1-score","ROC-AUC"],
  "Decision Tree":dt_metrics,
  "Random Forest":rf_metrics
})

print(metrics_df)
```

# <font color="magenta"> **Step 6: Visualize Performance Comparison**

```python
##Visualize the performance metrics using stacked bar plot
##~~~~~Write codes below~~~~~~~~~~
metrics_df.set_index("Metric").plot(kind="bar")
plt.title("Model Performance Comparison")
plt.ylabel("Score")
plt.grid(True)
plt.show()
```

# <font color="magenta"> **Step 7: ROC curve comparison**

```python
##Visualize ROC curves for both the methods on the same plot
##Use separate legends for the two models
##~~~~~Write codes below~~~~~~~~~~
fpr_dt, tpr_dt, _ = roc_curve(y_test, y_prob_dt)
fpr_rf, tpr_rf, _ = roc_curve(y_test, y_prob_rf)

plt.figure()

plt.plot(fpr_dt, tpr_dt, label="Decision Tree")
plt.plot(fpr_rf, tpr_rf, label="Random Forest")

plt.plot([0,1],[0,1],'k--')

plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve Comparison")
plt.legend()
plt.grid()

plt.show()
```

# <font color="magenta"> **Step 8: Display Confusion Matrics of the two models**

```python
##Calculate confusion matrics of DT and RF models
##Visualize confusion metrics
##~~~~~Write codes below~~~~~~~~~~
cm_dt = confusion_matrix(y_test, y_pred_dt)
cm_rf = confusion_matrix(y_test, y_pred_rf)

plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
sns.heatmap(cm_dt, annot=True, fmt='d', linewidth=.5, cmap='coolwarm')
plt.title("Decision Tree Confusion Matrix")

plt.subplot(1,2,2)
sns.heatmap(cm_rf, annot=True, fmt='d', linewidth=.5, cmap='Greens')
plt.title("Random Forest Confusion Matrix")

plt.show()
```

# <font color="magenta"> **Step 9: Visualize learned DT model decision boundary using PCA (2 Components) of training data**

```python
##Apply PCA on training set and project data in 2D-space
##Call PCA() function and set for 2 components and apply PCA on training data
##Train DT Classifier on PCA features on training set using Optimal_DT_Params
##Create 2D sample grid points for testing
##Set range of features using minimum and maximum value of PCA features
##Create mesh grid
##Predict class labels of grid points using trained DT model in 2D
##Visualize decision boundary of learned DT model using contourf()
##Scatter plot training samples in 2D with legends
##~~~~~Write codes below~~~~~~~~~~
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_train_pca = pca.fit_transform(X_train)

dt_pca = DecisionTreeClassifier(**Optimal_DT_Params)
dt_pca.fit(X_train_pca, y_train)

x_min, x_max = X_train_pca[:,0].min()-1, X_train_pca[:,0].max()+1
y_min, y_max = X_train_pca[:,1].min()-1, X_train_pca[:,1].max()+1

step = (x_max - x_min) / 1000   # ~30x30 grid ≈ 900 points

xx, yy = np.meshgrid(np.arange(x_min, x_max, step),
                     np.arange(y_min, y_max, step))

Z = dt_pca.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.contourf(xx, yy, Z, alpha=0.3)

plt.scatter(X_train_pca[:,0], X_train_pca[:,1], c=y_train)

plt.title("Decision Boundary (Decision Tree)")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")

plt.show()
```

# <font color="magenta"> **Step 10: Visualize learned RF model decision boundary using PCA (2 Components) of training data**

```python
##Train RF Classifier on PCA features on training set using Optimal_RF_Params
##Create 2D sample grid points for testing
##Set range of features using minimum and maximum value of PCA features
##Create mesh grid
##Predict class labels of grid points using trained RF model in 2D
##Visualize decision boundary of learned RF model using contourf()
##Scatter plot training samples in 2D with legends
##~~~~~Write codes below~~~~~~~~~~
rf_pca = RandomForestClassifier(**Optimal_RF_Params)
rf_pca.fit(X_train_pca, y_train)

Z = rf_pca.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.figure()
plt.contourf(xx, yy, Z, alpha=0.3)
plt.scatter(X_train_pca[:,0], X_train_pca[:,1], c=y_train)
plt.title("Decision Boundary (Random Forest)")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.show()
```
