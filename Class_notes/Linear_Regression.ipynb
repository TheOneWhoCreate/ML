# <font color="green">**1. Theory of Linear Regression**

Linear Regression is one of the simplest and most widely used supervised machine learning algorithms for predicting a continuous numerical value based on one or more input variables.

It models the relationship between a dependent variable (target) and one or more independent variables (features) by fitting a straight line to the observed data.

Linear regression assumes that there is a linear relationship between the input variables and the output variable.

# <font color="green">**2. Types of Linear Regression**

There are two main types of Linear Regression:

**a) Simple Linear Regression**

Simple Linear Regression involves one independent variable (x) and one dependent variable (y).

It is represented by the equation:

$𝑦=a𝑥+b$\
where, y = predicted output

$x$ = input feature

$a$ = slope of the line

$b$ = intercept

The goal of the model is to find the optimal values of
$a$ and $b$ that minimize prediction error.

These coefficients are obtained by using the Least Squares Method, which minimizes the total squared error between actual and predicted values.

## Formula for Slope $a$:

$$
\hat{a} =
\frac{
\frac{1}{n}\sum_{i=1}^{n} x_i y_i - \bar{x}\,\bar{y}
}{
\frac{1}{n}\sum_{i=1}^{n} x_i^2 - \bar{x}^2
}
=
\frac{\text{COV}(X,Y)}{\text{V}(X)}
=
r_{xy}\,\frac{\text{SD}(Y)}{\text{SD}(X)}
$$

##Formula for Intercept $b$:​
$$ \hat b = \bar{y} - \hat a\, \bar{x} $$



# <font color="green"> **3. Estimation of Linear Regression Coefficients and Fitting Line Step by Step from Scratch**

##<font color="blue"> **Problem-1:**</font> Given the following set of paired observations between an independent variable $x$ and a dependent variable $y$:

 | x | y  |
 | --| -- |
 | 1 | 2  |
 | 2 | 3  |
 | 3 | 5  |
 | 4 | 7  |
 | 5 | 11 |

Write a Python program to:

i) Compute the slope $a$ and intercept $b$ of the simple linear regression model using the least squares method.

ii) Obtain the regression equation of the form $y=ax+b$.

iii) Plot the original data points on a scatter plot and superimpose the fitted regression line on the same plot.

iv) Display the values of the estimated coefficients on the output.

```python
##Problem-1.(i): Computation of slope and intercept
##Step-1
##Import required Libraries
##============================
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# -------- Step 2: Load data set --------
##Construct/Declare feature vector and target variable
##===========================================
## ~~~ Write two lines of code here ~~~
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 3, 5, 7, 11])

##Count the number of samples
n = len(x)

##-------- Step 3: Compute means --------
##===========================================
## ~~~ Write two lines of code here ~~~
x_bar = np.mean(x)
y_bar = np.mean(y)

##------Step 4: Compute slope (â) using formula-------
##===========================================
## ~~~ Write one line of code here ~~~
a_hat = ((1/n) * np.sum(x * y) - (x_bar * y_bar)) / ((1/n) * np.sum(x**2) - (x_bar**2))

##--------Step 5: Compute intercept (b̂) --------
##===========================================
## ~~~ Write one line of code here ~~~
b_hat = y_bar - a_hat * x_bar


##Print the estimated coefficients:
print("Estimated coefficients:")
print(f"â (slope)     = {a_hat:.4f}")
print(f"b̂ (intercept) = {b_hat:.4f}")

##--------Step 6: Generate fitted line--------
##===========================================
## Problem 1.(ii)- Express the regression equation y=ax+b
## ~~~ Write one line of code here ~~~
print(f"Regression Equation: y = {a_hat:.4f} x + {b_hat:.4f}")

##-------- Step 7: Plot data and regression line --------
##===========================================
##Problem 1.(iii)-Scatter plot of the data and target
## ~~~ Write one line of code here ~~~
plt.scatter(x, y, label="Observed Data")

##===========================================
##Compute predicted values, plot fitted regression line by
##plotting observed vs. predicted values
## ~~~ Write two lines of code here ~~~
y_pred = a_hat * x + b_hat
plt.plot(x, y_pred, color='red', label="Fitted Line")

##Set labels, titles, etc.
plt.xlabel("X", fontsize=12, fontweight='bold')
plt.ylabel("Y", fontsize=12, fontweight='bold')
plt.title("Simple Linear Regression (Manual Computation)", fontsize=12, fontweight='bold')

eq_text = f"y = {a_hat:.4f}x + [{b_hat:.4f}]"
plt.text(0.40, 0.88, eq_text, transform=plt.gca().transAxes, fontsize=10, fontweight='bold')
plt.legend(fontsize=12, prop={'weight': 'bold'},bbox_to_anchor=(0.05, 1), loc='upper left')
plt.grid(True)
plt.show()
```

# <font color="green"> **5. Multivariate Linear Regression**

**Multivariate Linear Regression** involves more than one independent variable.

For example predicting **house price** based on size, number of rooms, and location.
A typical **housing dataset** contains one target variable (house price) and multiple predictor variables, such as:

|   Predictor (X)  | Meaning          |
| ---------------- | -------------------- |
| Area (sq ft)     | Size of house        |
| No. of bedrooms  | Rooms                |
| No. of bathrooms | Facilities           |
| Location score   | Neighborhood quality |
| Age of house     | Years since built    |
| Distance to city | Accessibility        |

For this problem the model learns an equation of the form:
$\hat {y}= β_0+β_1X_1 + β_1X_1+ .....+ β_6X_6$

or $\hat{y} = \beta_0 + \sum_{j=1}^{p} X_j \beta_j$

where

* $ \hat{𝑦} $  = predicted house price

* $𝑋_j$  = housing features

* $𝛽_j$  = learned coefficients

The coefficients are estimated by minimizing the Ordinary Least Squares (OLS) objective function:

$\min_{\beta} \; \sum_{i=1}^{n} (y_i - \hat{y}_i)^2$

or in vector form:

$\min_{\beta} \; ||Y - X\beta||^2 $

## <font color="green"> **5.1 Limitation of Ordinary Linear Regression**

**Problems with OLS:**

* Sensitive to multicollinearity

* Overfitting when number of features is large

* High variance in coefficient estimates

```python
##Problem-3.(i)
##Step 1: Import required libraries and
## Load housing dataset from sklearn.datasets
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_california_housing

##Load dataset
data = fetch_california_housing()

##Convert data to DataFrame
df = pd.DataFrame(data.data, columns=data.feature_names)
df['Price'] = data.target

##Check the features and target of the dataset
df.head()
```

```python
##Step 2: Separate X (features) and y (target) using df
##=====================================================
## ~~~ Write two lines of code here ~~~
X = df.drop(columns= ['Price'])
y = df['Price']
print(X.head())
print("\n")
print(y.head())
```

```python
##Step 3: Divide the dataset into training and testing set
##using train_test_split() function of sklearn.model_selection
from sklearn.model_selection import train_test_split

##Create stratification bins for the continuous target variable
##(divide house prices into 5 groups to apply stratified sampling)
y_binned = pd.qcut(y, q=5, labels=[0,1,2,3,4])

##Split the data set with 20% samples in testing set
X_train, X_test, y_train, y_test = train_test_split(
    X, y,test_size=0.2, shuffle=True, stratify=y_binned)
```

```python
##Step-4: Normalize features and targets within [0, 1]
##Import MinMaxScaler
from sklearn.preprocessing import MinMaxScaler
feature_scaler = MinMaxScaler()

## -------- Scale Features --------
##Fit the traing set features into the scaler
feature_scaler.fit(X_train)
##Scale training and Testing features using the trained feature_scaler
##=====================================================
## ~~~ Write two lines of code here ~~~
X_train_scaled = feature_scaler.transform(X_train)
X_test_scaled = feature_scaler.transform(X_test)

## -------- Scale Targets --------
target_scaler = MinMaxScaler()

##Convert Series to NumPy array before reshaping
y_train = y_train.values.reshape(-1, 1)
y_test = y_test.values.reshape(-1, 1)

## Fit scaler on training targets only
target_scaler.fit(y_train)

##Transform both train and test targets using the trained target_scaler
##=====================================================
## ~~~ Write two lines of code here ~~~
y_train_scaled = target_scaler.transform(y_train)
y_test_scaled = target_scaler.transform(y_test)
```

```python
##Step 5: Train Multivariate Linear Regression Model
##=====================================================
##Import Linear Regression model
from sklearn.linear_model import LinearRegression

##Define Linear Regression model and fit data to it
##=====================================================
## ~~~ Write two lines of code here ~~~
LR_model = LinearRegression()
LR_model.fit(X_train_scaled, y_train_scaled)

##Step 6: Print regression coefficients
coef_df = pd.DataFrame({
    'Feature': X.columns,
    'Coefficient': LR_model.coef_.ravel()
})

print("LR Model Intercept:", LR_model.intercept_)
print("LR Model coefficients:\n",coef_df)
```

```python
##Problem- 3.(ii)
##Step 7: Evaluation of Model Performance
##Import performance metrices
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

##Predictions on scaled training and Testing datasets
##=====================================================
## ~~~ Write two lines of code here ~~~
y_train_pred = LR_model.predict(X_train_scaled)
y_test_pred = LR_model.predict(X_test_scaled)


##Apply inverse transforms to calculate errors...
##Reshape the training and testing predictions
y_train_pred = y_train_pred.reshape((len(y_train_pred), 1))
y_test_pred = y_test_pred.reshape((len(y_test_pred), 1))

##Perform inverse transform
y_train_pred_lr = target_scaler.inverse_transform(y_train_pred)
y_test_pred_lr = target_scaler.inverse_transform(y_test_pred)

##Compute Training performance metrics and print the results
##=====================================================
## ~~~ Write three lines of code here ~~~
trn_mse_lr = mean_squared_error(y_train, y_train_pred_lr)
trn_mae_lr = mean_absolute_error(y_train, y_train_pred_lr)
trn_r2_lr = r2_score(y_train, y_train_pred_lr)

print("Performance on Training dataset:")
print("Training MSE:", trn_mse_lr)
print("Training MAE:", trn_mae_lr)
print("Training R²:", trn_r2_lr)

##Compute Testing performance metrics and print the results
##=====================================================
## ~~~ Write three lines of code here ~~~
tst_mse_lr = mean_squared_error(y_test, y_test_pred_lr)
tst_mae_lr = mean_absolute_error(y_test, y_test_pred_lr)
tst_r2_lr = r2_score(y_test, y_test_pred_lr)

print("Performance on Testing dataset:")
print("Testing MSE:", tst_mse_lr)
print("Testing MAE:", tst_mae_lr)
print("Testing R²:", tst_r2_lr)
```

##<font color="Green"> **6. Ridge Regression (L2 Regularization)**</font>

**Objective Function:**

Ridge regression modifies the OLS loss function by adding an L2 penalty term:

$\min_{\beta} \left\{ \sum_{i=1}^{n} (y_i - \hat{y}_i)^2 +
\alpha \sum_{j=1}^{p} \beta_j^2 \right\}$

or equivalently,

$\min_{\beta} \; ||Y - X\beta||^2 + \lambda ||\beta||_2^2$

where:

$\alpha$ = regularization parameter

$||\beta||_2^2 = \sum_{j=1}^{p} \beta_j^2 $

**Closed-form Solution:**

Ridge regression has an analytical solution:

$\hat{\beta}_{ridge} = (X^T X + \alpha I)^{-1} X^T Y$

where:

* $I$ is identity matrix

* $\alpha I$ stabilizes inversion and reduces variance

**Effect of Ridge Regularization:**

* Shrinks coefficients toward zero

* Does not make them exactly zero

* Reduces model variance

* Useful when features are highly correlated

##<font color="Green"> **7. Lasso Regression**</font>

**Objective Function:**
Lasso regression adds an L1 penalty instead of L2:

$\min_{\beta} \left\{ \sum_{i=1}^{n} (y_i - \hat{y}_i)^2 +
\alpha \sum_{j=1}^{p} |\beta_j| \right\}$

In vector form:

$\min_{\beta} \; ||Y - X\beta||^2 + \alpha ||\beta||_1$

where

$||\beta||_1 = \sum_{j=1}^{p} |\beta_j|$

**Effect of Lasso Regularization:**

* Shrinks some coefficients exactly to zero

* Performs automatic feature selection

* Produces sparse models

* Useful when many features are irrelevant


**How Lasso is Solved:**

* Unlike Ridge, Lasso has no closed-form solution. It is solved using:

* Coordinate Descent

* Gradient Descent

* Least Angle Regression (LARS)

These are implemented internally in sklearn.linear_model.Lasso.
